### Creating a function that does the precleaning part:describe,isnull(),info,checks duplicates 

I have decided to encapsulate the pre-cleaning process in a function for simplicity and eficiency.The goal of building a function for the pre-cleaning section is to streamline and automate the process of acquiring crucial dataset information. It enables us to assess data quality rapidly, find missing values and duplicates, and comprehend the overall structure of the dataset. We can easily apply these processes to various datasets and ensure consistency in our research by encapsulating them in a function.


```python
def pre_cleaning(df):
    """
    Perform pre-cleaning steps on the dataset.
    """
    # Display summary statistics
    print("Summary Statistics:")
    print(df.describe())
    print()

    # Check for missing values
    print("Missing Values:")
    print(df.isnull().sum())
    print()

    # Display dataset information
    print("Dataset Information:")
    print(df.info())
    print()

    # Check for duplicates
    duplicate_rows = df.duplicated().sum()
    print("Number of Duplicate Rows:", duplicate_rows)
    print()
```

```python
# Joint Analysis:We may combine the information from average ratings and the number of votes to create a 
# more thorough understanding. 
# Group the data by genre and calculate average rating and total number of votes
grouped = df.groupby('genres').agg({'averagerating': 'mean', 'numvotes': 'sum'}).reset_index()
grouped.head()
```

```python
#I want to find out overally with a combination of both votes and averagerating which genres are performing
#well.So i will incorporate some statistic technique called composite score by combining the average rating and 
#the number of votes using a weighted average.
# Calculate the composite score by combining the average rating and number of votes and adding new column to
#grouped
grouped['composite_score'] = grouped['averagerating'] * grouped['numvotes']
# Sort the grouped data by the composite score in descending order
top_genres_combined = grouped.sort_values(by='composite_score', ascending=False)

# Get the top genres based on the composite score
top_genres_combined = top_genres_combined.head(10)[['genres', 'composite_score']]

top_genres_combined_df = top_genres_combined.reset_index(drop=True)
top_genres_combined_df
```

```python
grouped.head(5)# here shows i have added another column of composite score in the grouped dataframe
```

I wanted to find the top genres that perform well in terms of both average ratings and amount of votes in order to offer better educated suggestions to Microsoft. I chose to achieve this by combining the average rating and the number of votes into a single composite score. By calculating this composite score, I can gain a better insight of which genres are popular with audiences and have a large amount of votes.

The composite score combines two essential factors: the average rating given by viewers and the number of votes received by a film. We can generate a single score that measures the overall performance of a genre by giving weights to each factor.
Similarly, we add weights to the average rating and the number of votes in our study. As a result, genres with higher average ratings and more votes will have a higher composite score. We can find genres that are not only highly rated but also popular among viewers by combining these two characteristics.
How we can achieve this is weighting the average rating and the number of votes according to their value to you. Multiply the average rating by its weight, as well as the number of votes, and then add the weighted numbers for each genre. This will provide me with a composite score for each genre that represents its overall performance. Finally, I will rank the genres by composite score to determine the top genres in terms of both average rating and popularity.


```python
plt.figure(figsize=(10, 6))
sns.barplot(data=top_genres_combined_df, y='genres', x='composite_score', palette='cividis')
plt.xlabel('Composite Score')
plt.ylabel('Genres')
plt.title('Top Genres Based on Composite Score')
plt.tight_layout()
plt.show()
```

##### Uncovering Film Sentiment Patterns Using Lexicon-Based Analysis for Microsoft's New Movie Studio
One of the things I want to explore is the sentiment of movie reviews, which tells us whether the reviews are positive or negative.

To accomplish this, I'm employing a simple technique known as "Lexicon-Based Sentiment Analysis." It's an excellent strategy for beginners because it doesn't necessitate the creation of a sophisticated model from the ground up. Instead, I can utilize pre-built lexicons, which are similar to dictionaries in that they contain words as well as their associated sentiment scores (such as positive, negative, or neutral).


This is how it works: I'll go through the movie reviews and cross-reference each word with the dictionary to determine whether it's favorable, negative, or neutral. Then, for each review, I'll compute an overall sentiment score.


I can acquire insights into the general sentiment of movie reviews by running this sentiment analysis. Positive sentiment numbers may suggest that audiences appreciated the films, however low sentiment levels may identify places for improvement. This information will be crucial to Microsoft's new film studio, as it will help them understand how consumers feel about different films and guide their decisions on what types of films to develop in order to optimize popularity and appeal to their target audience.



```python
df.head()
```

First, I require the nltk package, which has essential tools for natural language processing (NLP). NLP is a method for computers to comprehend and interact with human language. The nltk library assists me with tasks like as tokenizing words (splitting text into individual words), accessing stopwords (frequently used words such as "the" or "and"), and more.


Following that, I'm employing the Afinn library, which has a vocabulary of words as well as their related emotion scores. Sentiment scores indicate if a word's meaning is good, negative, or neutral. For instance, "happy" may have a positive score, whilst "sad" may have a negative value.


I'm also importing the nltk stopwords corpus, which is a collection of common words that we don't want to examine when analyzing text. These words, such as "and" or "is," add little meaning to the analysis.

I can do lexicon-based sentiment analysis using these packages. This means I can look through movie reviews, break them down into individual words, use the Afinn lexicon to evaluate their sentiment scores, and determine whether the reviews are typically good or negative. It's a simple and beginner-friendly technique to learning about how people feel about movies without having to develop a complex machine learning model.


```python
import nltk#this is a library that aids in natural language processing activities such as splitting text 
#into words and accessing stopwords.
from afinn import Afinn#This is a library with a dictionary of words that have sentiment scores. 
#It assists me in determining whether words have a good, negative, or neutral connotation.
from nltk.corpus import stopwords#This is a collection of frequent terms that we often want to avoid when 
#studying text since they contribute little meaning.
from nltk.tokenize import word_tokenize#This is a nltk function that tokenizes text into individual words 
#for subsequent analysis.
```

```python
#I changed the movie titles to lowercase after removing any special characters (such as punctuation). 
#This ensures that all titles are uniform and easy to deal with.
title_column = 'primary_title'
#In the 'primary_title' column, I handled any missing values. To guarantee we have complete and 
#legitimate data, I eliminated the entire row from the DataFrame if a movie title was missing.
df[title_column] = df[title_column].str.replace('[^\w\s]', '').str.lower()
# I will Handle missing values
df.dropna(subset=[title_column], inplace=True)
#These procedures ensure that the film titles are clean, consistent, and ready for further examination.
```

```python
#I'm using the NLTK package to get a list of common English stopwords, which are words with little 
#significance that are frequently deleted during text analysis. Then I wrote a method called 
#remove_stopwords that accepts as input a string (movie title).
nltk.download('stopwords')
nltk.download('punkt')

stop_words = set(stopwords.words('english'))
#I tokenize the text into individual words within the function and then filter out any terms that 
#appear in the list of stopwords. This implies that I'm deleting terms like "the", "and", "is", and so on,
#which are prevalent in English but don't add anything to the sense of the movie titles.

def remove_stopwords(text):
    words = word_tokenize(text)
    filtered_words = [word for word in words if word.lower() not in stop_words]
    return ' '.join(filtered_words)
#Finally, I apply this remove_stopwords function to the DataFrame (df) column 'primary_title'. 
#By doing so, I remove stopwords from all movie titles, making them more useful for subsequent analysis and minimizing data noise.
df[title_column] = df[title_column].apply(remove_stopwords)
```

```python
afinn = Afinn()
#To calculate sentiment scores for each movie title in the DataFrame (df), I'm using the Afinn library, 
#which is a lexicon-based sentiment analysis tool.

Afinn contains a built-in dictionary of words with accompanying sentiment evaluations (positive or negative). I wrote a function called get_sentiment_score that accepts a movie title (text) as input and returns the sentiment score for that title.
# I then apply this function to the DataFrame's 'primary_title' column, and it generates the sentiment
#score for each movie title based on the words in the title and their related Afinn lexicon scores.

# This gives me a sentiment score for each movie title, which can tell us if the title is seen as more 
#positive or negative. 
def get_sentiment_score(text):
    return afinn.score(text)
This information may be valuable for Microsoft's new film studio in understanding 
#the prospective audience perception of various movie titles and assisting them in making educated 
#decisions about the types of films they wish to create.
df['sentiment_score'] = df[title_column].apply(get_sentiment_score)
```

```python
#Lets now visualize our output .The sentiment scores we calculated using the Afinn library, and the 
#histogram is a graphical representation that shows how many movie titles fall into different sentiment 
#score ranges.
plt.hist(df['sentiment_score'], bins=20, edgecolor='k')
plt.xlabel('Sentiment Score')
plt.ylabel('Number of Titles')
plt.title('Distribution of Sentiment Scores');
```

In this sentiment score histogram, I can see that the majority of movie titles have a sentiment score of 0, indicating a neutral sentiment. As a result, the majority of movie names are neither strongly positive nor strongly negative.

A considerable proportion of movie titles have negative sentiment scores, particularly -2, -3, -1, -4, and -5. This implies that a significant number of movie names contain negative implications or sentiments in their phrasing.

On the plus side, a few movie names have sentiment scores of 3, 2, 1, 4, and 5, indicating that they have positive wording that may appeal to moviegoers.


However, several outliers stand out, such as emotion scores of 36, -12, -10, -9, and -11. These are extremely high sentiment scores that should be investigated further because they could reveal data problems or unusual cases.
